# Visualization with Matplotlib

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure() 
plt.plot(x, np.sin(x), '-') 
plt.plot(x, np.cos(x), '--');
```

```python
plt.figure() # create a plot figure
# create the first of two panels and set current axis 
plt.subplot(2, 1, 1) # (rows, columns, panel number) 
plt.plot(x, np.sin(x))
# create the second panel and set current axis
plt.subplot(2, 1, 2) 
plt.plot(x, np.cos(x));
```

```python
 # First create a grid of plots
# ax will be an array of two Axes objects 
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x)) 
ax[1].plot(x, np.cos(x));
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure() 
ax = plt.axes()
x = np.linspace(0, 10, 1000) 
ax.plot(x, np.sin(x));
```

## Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue') 
plt.plot(x, np.sin(x - 1), color='g') 
plt.plot(x, np.sin(x - 2), color='0.75')
```

```python
plt.plot(x, x + 0, linestyle='solid') 
plt.plot(x, x + 1, linestyle='dashed') 
plt.plot(x, x + 2, linestyle='dashdot') 
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes: 
plt.plot(x, x + 4, linestyle='-') # solid 
plt.plot(x, x + 5, linestyle='--') # dashed 
plt.plot(x, x + 6, linestyle='-.') # dashdot 
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
plt.plot(x, x + 0, '-g') # solid green 
plt.plot(x, x + 1, '--c') # dashed cyan 
plt.plot(x, x + 2, '-.k') # dashdot black 
plt.plot(x, x + 3, ':r'); # dotted red
```

## Axes Limits

```python
plt.plot(x, np.sin(x)) 
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x)) 
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x)) 
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x)) 
plt.axis('equal')
```

## Labeling Plots

```python
plt.plot(x, np.sin(x)) 
plt.title("A Sine Curve")
plt.xlabel("x") 
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') 
plt.plot(x, np.cos(x), ':b', label='cos(x)') 
plt.axis('equal')
plt.legend()
```

## object-oriented interface to plotting

```python

ax = plt.axes() 
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', 
       title='A Simple Plot');
```

## Scatter Plots with plt.plot

```python
x = np.linspace(0, 10, 30) 
y = np.sin(x)
z = y + 1
plt.plot(x, y, 'o', color='g', label="y")
plt.plot(x, z, '-ok', label='z')
plt.legend()
```

```python
plt.plot(x, y, '-D', color='r', markersize=10, linewidth=3, 
         markerfacecolor='g',
         markeredgecolor='y',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

## Scatter Plots with plt.scatter
<span style="color:green">plt.scatter has the capability to render a different size and/or color for each point</span>

```python
rng = np.random.RandomState(0) 
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
plt.ylim(-3, 3)
```

## Histograms

```python
x1 = np.random.normal(0, 0.8, 1000) 
x2 = np.random.normal(-2, 1, 1000) 
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.5, bins=40)
plt.hist(x1, **kwargs) 
plt.hist(x2, **kwargs) 
plt.hist(x3, **kwargs)
```

```python
data = np.random.randint(1000, size=100)
counts, bin_edges = np.histogram(data, bins=5)
print(data)
print(counts)
print(bin_edges)
```

## Two-Dimensional Histograms and Binnings

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
x.shape
y.shape
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues') 
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues') 
cb = plt.colorbar(label='count in bin')
```

```python
from scipy.stats import gaussian_kde
data = np.vstack([x, y])
kde = gaussian_kde(data)
data.shape
```
